```python

# Simple Keyword-Based AI Assistant
def assistant():
    print("AI: Hello! I am your Study Assistant. Ask me about 'Math', 'Physics', or 'AI'.")
    user_input = input("You: ").lower()

    responses = {
        "math": "Mathematics is the language of AI. Focus on Linear Algebra and Calculus!",
        "physics": "Physics helps us understand how AI can interact with the real world.",
        "ai": "Artificial Intelligence is the future! You are on the right track."
    }

    print("AI:", responses.get(user_input, "That's interesting! I am still learning about that."))

assistant()
```
